```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros
s = 20
r = 3

# Visualización (usando matplotlib)
plt.figure(figsize=(6, 6))
plt.plot([0, s], [0, s], 'k-')  # Marco del cuadrado
plt.gca().set_aspect('equal', adjustable='box')
circle = plt.Circle((s / 2, s / 2), r, color='r', fill=False)
plt.gca().add_patch(circle)
plt.xlim(0, s)
plt.ylim(0, s)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Probabilidad Geométrica: Punto en un Área")
plt.text(s / 2, s * 0.9, f"Área Cuadrado = {s**2}", ha='center', va='center')
plt.text(s / 2, s * 0.1, f"Área Círculo = {np.round(np.pi * r**2, 2)}", ha='center', va='center', color='red')

# Probabilidad
probabilidad = np.round((np.pi * r**2) / (s**2), 2)
print(f"P(A) = {probabilidad}")
plt.show()


# Problema de los amigos

# Espacio muestral (cuadrado de 60x60)
total_area = 60 * 60

# Área favorable (región donde |X-Y| <= 15)
favorable_area = total_area - 2 * ((60 - 15)**2 / 2)

# Probabilidad
probability_meeting = favorable_area / total_area
print(f"\nProbabilidad de que se encuentren: {probability_meeting}")


# Representación gráfica (usando matplotlib)
x = np.linspace(0, 60, 100)
y1 = x + 15
y2 = x - 15

plt.figure(figsize=(6, 6))
plt.plot(x, y1, 'r--', label='|X-Y|=15')
plt.plot(x, y2, 'r--')
plt.fill_between(x, y1, 60, where=(y1 < 60), color='skyblue', alpha=0.5)  # Zona por encima de y1
plt.fill_between(x, 0, y2, where=(y2 > 0), color='skyblue', alpha=0.5)  # Zona por debajo de y2
plt.xlim(0, 60)
plt.ylim(0, 60)
plt.xlabel("Tiempo de llegada del primer amigo (minutos)")
plt.ylabel("Tiempo de llegada del segundo amigo (minutos)")
plt.title("Probabilidad de encuentro")
plt.grid(True)
plt.legend()
plt.show()
```
